```python
# Defining the F1 Score Metric
def f1_score(y_true, y_pred):
    # Convert predictions to binary (0 or 1) using a threshold of 0.5
    y_pred_binary = K.round(y_pred)

    # Calculate true positives, false positives, and false negatives
    tp = K.sum(K.cast(y_true * y_pred_binary, 'float'), axis=0)    # true positives
    fp = K.sum(K.cast((1 - y_true) * y_pred_binary, 'float'), axis=0) # false positives
    fn = K.sum(K.cast(y_true * (1 - y_pred_binary), 'float'), axis=0) # false negatives

    # Precision and recall
    precision = tp / (tp + fp + K.epsilon())
    recall = tp / (tp + fn + K.epsilon())

    # Compute F1 score
    f1 = 2 * (precision * recall) / (precision + recall + K.epsilon())
    return K.mean(f1)
```

```python
def binary_classifier(extracted_features, actual_outputs, num_hidden_layers, regularization_constant, num_neurons_per_layer, learning_rate):
  input_layer = keras.layers.Input(shape=(extracted_features.shape[1], ))
  n = num_hidden_layers
  hidden_layer = keras.layers.Dense(num_neurons_per_layer, activation='tanh', kernel_regularizer=regularizers.l2(regularization_constant))(input_layer)
  hidden_layer = layers.BatchNormalization()(hidden_layer)
  for i in range(n-1):
      hidden_layer = keras.layers.Dense(num_neurons_per_layer, activation='tanh', kernel_regularizer=regularizers.l2(regularization_constant))(hidden_layer)
      hidden_layer = layers.BatchNormalization()(hidden_layer)

  output_layer = keras.layers.Dense(1, activation='sigmoid')(hidden_layer)

  model = keras.models.Model(inputs=input_layer, outputs=output_layer)
  model.compile(optimizer=keras.optimizers.Adam(learning_rate=learning_rate), loss='binary_crossentropy', metrics=['accuracy',  'recall', lambda y_true, y_pred: f1_score(y_true, y_pred)])
  return model
```

```python
import optuna
from sklearn.model_selection import train_test_split

def objective(trial):
    # Define the hyperparameter search space
    num_layers = trial.suggest_int('num_layers', 3, 10)
    num_neurons = trial.suggest_int('num_neurons', 64, 256, step=64)
    learning_rate = trial.suggest_loguniform('learning_rate', 1e-5, 1e-1)
    regularization_constant = trial.suggest_loguniform('regularization_constant', 1e-5, 1e-1)


    # Train/validation split
    input_train_split, input_val_split, actual_outputs_train_split, actual_outputs_val_split = train_test_split(
        input_train, actual_outputs_train, test_size=0.1, random_state=42
    )

    # Build the model
    input_shape = input_train_split.shape[1]
    output_shape = actual_outputs_train_split.shape[1]
    input_layer = keras.Input(shape=(input_shape,))
    output_layer = keras.Input(shape=(output_shape,))

    model = binary_classifier(extracted_features= input_train, actual_outputs = actual_outputs_train, num_hidden_layers = 5, regularization_constant = regularization_constant, num_neurons_per_layer = num_neurons, learning_rate= learning_rate)
    # Train the model and evaluate on the validation set
    history = model.fit(
        input_train_split,
        actual_outputs_train_split,
        epochs=10,
        batch_size=64,
        validation_data=(input_val_split, actual_outputs_val_split),
        verbose=0
    )

    # Return the validation loss to minimize
    return history.history['val_loss'][-1]  # Return the last validation loss value

```

```python
# Create an Optuna study and optimize
study = optuna.create_study(direction='minimize')
study.optimize(objective, n_trials=25)

# Best parameters and loss
print("Best parameters:", study.best_params)
```

```python
best_params = study.best_params

best_num_layers = best_params['num_layers']
best_num_neurons = best_params['num_neurons']
best_learning_rate = best_params['learning_rate']
best_regularization_constant = best_params['regularization_constant']
print("\nbest_num_neurons:", best_num_neurons,
      "\nBest num_layers:", best_num_layers,
      "\nBest learning_rate:", best_learning_rate,
      "\nBest regularization_constant:", best_regularization_constant)
```

```python
damage_classification_model = binary_classifier(extracted_features, actual_outputs, num_hidden_layers = best_num_layers, regularization_constant = best_regularization_constant, num_neurons_per_layer = best_num_neurons, learning_rate = best_learning_rate)
damage_classification_model.summary()
```

```python
# Save the model
history = damage_classification_model.fit(input_train, actual_outputs_train, epochs=20, batch_size=32, verbose=1)
# model_save_path = '/content/drive/Othercomputers/My PC/GDrive/CEE286_Project/trained_models/damage_classification_model.keras'
model_save_path = '/content/drive/MyDrive/CEE286_Project/trained_models/damage_classification_model.keras'
damage_classification_model.save(model_save_path)
```
